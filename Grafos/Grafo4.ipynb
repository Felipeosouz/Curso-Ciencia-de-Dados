```python
from igraph import Graph
from igraph import plot

grafo = Graph(edges = [(0,1), (2,3), (0,2), (0,3)], directed = True)
grafo.vs["label"] = ["Felipe", "Carlos", "Vinicius", "Filipe"]
grafo.vs["peso"] = [40,30,30,25]
grafo.es["TipoAmizade"] = ["Amigo", "Inimigo", "Inimigo", "Amigo"]
grafo.es["weight"] = [1,2,1,3]

# Definição de cores para os vértices
grafo.vs["cor"] = ["pink", "purple", "orange", "cyan"]
plot(grafo, bbox = (0, 0, 250, 250), vertex_color = grafo.vs["cor"])
```

```python
# Peso para as arestas
plot(grafo, bbox = (250, 250), edge_width = grafo.es["weight"], 
    vertex_color = grafo.vs["cor"])
```

```python
# Pesos para os vértices
plot(grafo, bbox = (250, 250), vertex_size = grafo.vs["peso"],
    edge_width = grafo.es["weight"], 
    vertex_color = grafo.vs["cor"])
```

```python
# Curvatura
plot(grafo, bbox = (250, 250), vertex_size = grafo.vs["peso"],
    edge_width = grafo.es["weight"], 
    vertex_color = grafo.vs["cor"],
    edge_curved = 0.4)
```

```python
# Formato
plot(grafo, bbox = (250, 250), vertex_size = grafo.vs["peso"],
    edge_width = grafo.es["weight"], 
    vertex_color = grafo.vs["cor"],
    edge_curved = 0.4, vertex_shape = "triangle")
```
